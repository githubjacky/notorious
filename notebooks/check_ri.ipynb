```python
def gtab_res(query: str,
                   geo: str = "", 
                   period: List[str] = ["2016-01-01", "2018-07-31"],
                   init_path: str = "gtab_config"
                  ):
    t = GTAB(dir_path = init_path)
    anchor_dir = Path(f"{init_path}/output/google_anchorbanks")
    anchor_file = "_".join((
        f"google_anchorbank_geo={geo}",
        f"timeframe={period[0]} {period[1]}.tsv"
    ))
    anchor_path = anchor_dir / anchor_file
    if not anchor_path.exists():
        t.set_options(pytrends_config={
            "geo": "",
            "time_frame": " ".join(period)
        })
    t.create_anchorbank()
    t.set_active_gtab(anchor_file)
    logger.info(f'new_query: {query}')
    return t.new_query(query)
```

```python
query = 'Harvey Weinstein'
res = gtab_res(query)
res
```

## check pandas resampling meghod

```python
counts ={}
for i in [str(i).split('-')[:-1] for i in res.get('max_ratio').index]:
    key = '_'.join(i)
    if key not in counts: 
        counts[key] = 1
    else:
        counts[key] += 1
```

```python
agg_mean = []
begin = 0
end = 0
for val in counts.values():
    end += val
    agg_mean.append(res.get('max_ratio').iloc[beg:end].mean())
    beg = end

pd.DataFrame({
    'pd': res.get('max_ratio').resample('M').mean(),
    'manual': agg_mean
})
```

```python
agg_sum = []
begin = 0
end = 0
for val in counts.values():
    end += val
    agg_sum.append(res.get('max_ratio').iloc[beg:end].sum())
    beg = end

pd.DataFrame({
    'pd': res.get('max_ratio').resample('M').sum(),
    'manual': agg_sum
})
```
